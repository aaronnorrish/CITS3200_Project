This script will open the merged_list xlsx file and parse the journal names into a lowercase format as well as return the URL to get to the home page, and submissions page.

```python
import pandas as pd
import openpyxl as xl
import pprint as pp
```

```python
#Load xlsx file into Python
wb = xl.load_workbook('Merged_List.xlsx') #Open Input workbook
journals = wb['journals'] #Open journals worksheet
urls = wb["Sheet2"]
```

```python
#Add URL access method to 'journals' worksheet
journals['H1'].value = "home_page"

access_methods = {}
publishers = []
url_methods = []

for row in urls.iter_rows(min_row=2, max_col=5, max_row=17):
    publishers.append(row[0].value)
    url_methods.append(row[4].value) 
    
access_methods = dict(zip(publishers,url_methods))
pp.pprint(access_methods)
```

```python
#Iterate through publisher names and append url access method. 
#Then make lowercase + replace spaces with hyphens
#removes hyphens from issn and eissns
#Its innefficient of course to use so many loops, but this can be changed later if necassary

row_count = 2 #Starts at first cell
for row in journals.iter_rows(min_row=2, min_col=2, max_col=2, max_row=5035):#Iterate Publishers
    for publisher in row:
        for key, value in access_methods.items():
            if key == publisher.value:
                journals['H'+str(row_count)].value = access_methods[publisher.value]
        row_count += 1
        
            
for row in journals.iter_rows(min_row=2, max_col=4, max_row=5035): #Iterate Journals
    for journal in row:
        row[0].value = row[0].value.lower()
        row[0].value = row[0].value.replace(" ", "-")# Replace space with hyphen
        row[2].value = row[2].value.replace("-", "") # Remove hyphen from ISSN
        if row[3].value != None:
            row[3].value = row[3].value.replace("-", "") # Remove hyphen from EISSN
        
```

```python
#Iterates through urls and concatontes appropriate information into file
#eg journal name = psychology-review
# old url = www.journal/JNAME.com
# new url = www.journal/psychology-review.com

row_count = 2 #Starts at first Cell
for row in journals.iter_rows(min_row=2, max_row=5035, min_col=8, max_col=8):#Iterate over URLS
    for url in row:
        if "JNAME" in str(url.value):
            url.value = url.value.replace("JNAME",str(journals['A'+str(row_count)].value)) #Replace variable with journal name
        if "EISSN" in str(url.value):
            url.value = url.value.replace("EISSN",journals['D'+str(row_count)].value)#Replace variable with EISSN
        row_count+= 1
        
        
```

```python
#Add URL access method to 'journals' worksheet
journals['I1'].value = "submission_page"

access_methods = {}
publishers = []
url_methods = []

for row in urls.iter_rows(min_row=2, max_col=8, max_row=17):
    publishers.append(row[0].value)
    url_methods.append(row[7].value) 
    
access_methods = dict(zip(publishers,url_methods))
pp.pprint(access_methods)
```

```python
row_count = 2 #Starts at first cell
for row in journals.iter_rows(min_row=2, min_col=2, max_col=2, max_row=5035):#Iterate Publishers
    for publisher in row:
        for key, value in access_methods.items():
            if key == publisher.value:
                journals['I'+str(row_count)].value = access_methods[publisher.value]
        row_count += 1
```

```python
row_count = 2 #Starts at first Cell
for row in journals.iter_rows(min_row=2, max_row=5035, min_col=9, max_col=9):#Iterate over URLS
    for url in row:
        if "JNAME" in str(url.value):
            url.value = url.value.replace("JNAME",str(journals['A'+str(row_count)].value)) #Replace variable with journal name
        if "EISSN" in str(url.value):
            url.value = url.value.replace("EISSN",journals['D'+str(row_count)].value)#Replace variable with EISSN
        if "ISSN" in str(url.value):
            url.value = url.value.replace("ISSN",journals['C'+str(row_count)].value)#Replace variable with ISSN
        row_count+= 1
        print(url.value)
        
```
